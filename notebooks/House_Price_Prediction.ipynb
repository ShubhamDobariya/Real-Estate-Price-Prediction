### Data Preprocessing 

```python
## separate feature and target
X = df.drop(columns=[TARGET_COL])
y = df[TARGET_COL]
```

```python
X.head()
```

```python
y.head()
```

```python
## train test split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=RANDOM_STATE
)
```

```python
print("Train shape", X_train.shape)
print("Test shape", X_test.shape)
```

**Preprocessing Pipeline**

```python
numerical_features = X_train.select_dtypes(include=[np.number]).columns.tolist()
categorical_features = X_train.select_dtypes(exclude=[np.number]).columns.tolist()
```

```python
print("Numerical Features: ",numerical_features)
print("Categorical Features: ", categorical_features)
```

```python
## numerical features - preprocessing steps
numerical_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler())
    ]
)
```

```python
## categorical features - preprocessing steps
categorical_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("onehot", OneHotEncoder(handle_unknown="ignore"))
    ]
)
```

```python
## preprocessing pipeline
preprocess = ColumnTransformer(
    transformers=[
        ("num", numerical_transformer, numerical_features),
        ("cat", categorical_transformer, categorical_features),
    ]
)
```

**Baseline model (No CV, No Tuning)**

```python
baseline_pipe = Pipeline(
    steps=[
        ("preprocess", preprocess),
        ("model", LinearRegression())
    ]
)
```

```python
## preprocess the data and train the baseline model
baseline_pipe.fit(X_train, y_train)
```

**Evaluation of baseline model**

```python
train_baseline_model = baseline_pipe.predict(X_train)
test_baseline_model = baseline_pipe.predict(X_test)
```

```python
train_baseline_model[:5]
```

```python
y_train[:5]
```

```python
train_baseline_rmse = root_mean_squared_error(y_train, train_baseline_model)
train_baseline_mae = mean_absolute_error(y_train, train_baseline_model)
train_baseline_r2 = r2_score(y_train, train_baseline_model)
```

```python
print("===== TRAIN BASELINE MATRICS(LinearRegresion) ======")
print(f"RMSE: {train_baseline_rmse:.3f}")
print(f"MAE: {train_baseline_mae:.3f}")
print(f"R2_score: {train_baseline_r2:.3f}")
```

```python
test_baseline_rmse = root_mean_squared_error(y_test, test_baseline_model)
test_baseline_mae = mean_absolute_error(y_test, test_baseline_model)
test_baseline_r2 = r2_score(y_test, test_baseline_model)
```

```python
print("===== TEST BASELINE MATRICS(LinearRegresion) ======")
print(f"RMSE: {test_baseline_rmse:.3f}")
print(f"MAE: {test_baseline_mae:.3f}")
print(f"R2_score: {test_baseline_r2:.3f}")
```

**Model Selection & Optimization**

```python
models = {
    "LinearRegression":LinearRegression(),
    "Ridge": Ridge(random_state=RANDOM_STATE),
    "Lasso": Lasso(random_state=RANDOM_STATE,max_iter=10000),
    "RandForest": RandomForestRegressor(),
    "HistGB": HistGradientBoostingRegressor()
}
```

```python
k = 5
cv = KFold(n_splits=k, shuffle=True, random_state=RANDOM_STATE)
```

```python
scoring = {
    "rmse": "neg_root_mean_squared_error",
    "mae": "neg_mean_absolute_error",
    "r2": "r2"
}
```

```python
rows= []

for name, model in models.items():
    pipe = Pipeline(
        steps=[
            ("preprocess", preprocess),
            ("model", model)
        ]
    )
    
    scores = cross_validate(pipe, X_train, y_train, cv=cv, scoring=scoring, n_jobs=1)
    rows.append({
        "model": name,
        "cv_rmse": -scores["test_rmse"].mean(),
        "cv_mae": -scores["test_mae"].mean(),
        "cv_r2": scores["test_r2"].mean()
    })
    

## sort based on lowest rmse value
cv_results = pd.DataFrame(rows).sort_values("cv_rmse")
print("===== Model Comparison =====") 
print(cv_results)
```

```python
cv_results
```

```python
rows
```

```python
best_row = cv_results.sort_values("cv_rmse").iloc[0]

best_model = best_row["model"]
best_rmse = best_row["cv_rmse"]

print("Best model based on CV RMSE")
print("Model : ", best_model)
print("CV RMSE : ", best_rmse)
```

**Best Model : HistGradientBoostingRegressor**

**Hyperparameters Tuning**

```python
hgb_pipe = Pipeline(
    steps=[
        ("preprocess", preprocess),
        ("model", HistGradientBoostingRegressor(random_state=RANDOM_STATE))
    ]
)
```

```python
param_grid = {
    "model__learning_rate":[0.03, 0.05, 0.1],
    "model__max_depth": [None, 3, 6],
    "model__max_leaf_nodes":[15, 31, 63],
    "model__min_samples_leaf":[20, 50, 100],
    "model__l2_regularization":[0.0, 0.1, 1.0]
}
```

```python
grid = GridSearchCV(
    estimator=hgb_pipe,
    param_grid=param_grid,
    cv=cv,
    scoring="neg_root_mean_squared_error",
    n_jobs=-1,
    verbose=1
)
```

```python
grid.fit(X_train, y_train)
```

```python
print("===== TUNED HistGB (cv) =====")
print("Best CV RMSE: ", -grid.best_score_)
print("Best params: ", grid.best_params_)
```

classification -> accuracry score -> highest -> best model

regression -> error -> lowest -> best model

**Retraining with best params**

```python
hgb_best = Pipeline(
    steps=[
        ("preprocess", preprocess),
        ("model", HistGradientBoostingRegressor(
            l2_regularization=0.1, 
            learning_rate=0.1, 
            max_depth=None, 
            max_leaf_nodes=63, 
            min_samples_leaf=20
        ))
    ]
)
```

```python
## train best model on entier training data (can be done with refit=True in grid search)
hgb_best.fit(X_train, y_train)
```

**Final Evalution**

```python
train_final_pred = hgb_best.predict(X_train)

train_final_rmse = root_mean_squared_error(y_train, train_final_pred)
train_final_mae = mean_absolute_error(y_train, train_final_pred)
train_final_r2 = r2_score(y_train, train_final_pred)

print("===== FINAL MODEL(Tuned HGB) Train Performance ======")
print(f"RMSE: {train_final_rmse:.3f}")
print(f"MAE: {train_final_mae:.3f}")
print(f"R2_score: {train_final_r2:.3f}")
```

```python
test_final_pred = hgb_best.predict(X_test)

test_final_rmse = root_mean_squared_error(y_test, test_final_pred)
test_final_mae = mean_absolute_error(y_test, test_final_pred)
test_final_r2 = r2_score(y_test, test_final_pred)

print("===== FINAL MODEL(Tuned HGB) Test Performance ======")
print(f"RMSE: {test_final_rmse:.3f}")
print(f"MAE: {test_final_mae:.3f}")
print(f"R2_score: {test_final_r2:.3f}")
```

```python
## residual plot
residuals = y_test - test_final_pred


plt.figure(figsize=(6, 4))
plt.scatter(test_final_pred, residuals, s=10)
plt.axhline(0)
plt.title("Residual vs Prediction")
plt.xlabel("Predicted")
plt.ylabel("Residuals")
plt.show()
```
